### **Exercise 1**: Function Plotting and Analytical Determination of Critical Points

```python
# Import numpy and pyplot libraries
import numpy as np
import math 
import matplotlib.pyplot as plt
```

```python
def f1(x, y):
    return x**3 + y**3 - 3*x - 12*y

def f2(x, y):
    return x**3 - 12*x*y + 8*y**3

def f3(x, y):
    return (x - 1)**2 + 10 * (x**2 - y)**2
```

3.4 The instruction block below provides the template for the function d2f1, which calculates and returns the Hessian matrix of $f_1$. Complete this instruction block so that d2f1 does what is expected. Then use this function to verify the nature of the critical points of $f_1$.

```python
# Instruction block to be completed
# Hessian of f1
def d2f1(x, y):
    d2f_dxdx = 6*x
    d2f_dxdy = 0
    d2f_dydx = 0
    d2f_dydy = 6*y
    return np.array([[d2f_dxdx, d2f_dxdy],
                     [d2f_dydx, d2f_dydy]])

# Classify critical points using eigenvalues
for (x, y) in crit_f1:
    H = d2f1(x, y)
    eigs = np.linalg.eigvals(H)
    print(f"Point ({x},{y}) -> eigenvalues = {eigs}")
```

```python
# To be completed to determine the nature of the critical points of f1

# List of critical points of f1 found analytically
crit_f1 = [(-1, -2), (-1,  2), (1, -2), (1,  2)]

for (x, y) in crit_f1:
    # Calculate the Hessian matrix at (x, y)
    matrice = d2f1(x, y)
    
    # Calculate the eigenvalues of the Hessian matrix
    valProp = np.linalg.eigvals(matrice)
    
    print(f'Point ({x},{y})')
    print('  Hessian = {}'.format(matrice))
    print('  vp1 = {}'.format(valProp[0]))
    print('  vp2 = {}'.format(valProp[1]))
    
    # Determine the nature of the critical point
    if (valProp[0] > 0) and (valProp[1] > 0):
        nature = 'local minimum'
    elif (valProp[0] < 0) and (valProp[1] < 0):
        nature = 'local maximum'
    else:
        nature = 'saddle point'
        
    print('  Nature : {}\n'.format(nature))
```

3.5 Redo questions 3.2, 3.3, and 3.4 for the function $f_2$ from exercise 1.

```python
# Exercise 3.5 - Determine the nature of the critical points of f2

# Gradient of f2
def df2(x, y):
    df_dx = 3*x**2 - 12*y
    df_dy = -12*x + 24*y**2
    return np.array([df_dx, df_dy])

# Hessian of f2
def d2f2(x, y):
    d2f_dxdx = 6*x
    d2f_dxdy = -12
    d2f_dydx = -12
    d2f_dydy = 48*y
    return np.array([[d2f_dxdx, d2f_dxdy],
                     [d2f_dydx, d2f_dydy]])

# Critical points of f2 found analytically
crit_f2 = [(0, 0), (2, 1)]

for (x, y) in crit_f2:
    # Check gradient
    grad = df2(x, y)
    
    # Calculate the Hessian matrix at (x, y)
    matrice = d2f2(x, y)
    
    # Eigenvalues of the Hessian
    valProp = np.linalg.eigvals(matrice)
    
    print(f'Point ({x},{y})')
    print('  Gradient = {}'.format(grad))
    print('  Hessian = {}'.format(matrice))
    print('  vp1 = {}'.format(valProp[0]))
    print('  vp2 = {}'.format(valProp[1]))
    
    # Nature of the critical point
    det = np.linalg.det(matrice)
    tr = np.trace(matrice)
    if det > 0 and tr > 0:
        nature = 'local minimum'
    elif det > 0 and tr < 0:
        nature = 'local maximum'
    else:
        nature = 'saddle point'
        
    print('  Nature : {}\n'.format(nature))
```
